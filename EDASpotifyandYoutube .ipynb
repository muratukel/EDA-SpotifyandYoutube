# Finding Outliers

**GÖRÜNTÜLENMELER İÇİN AYKIRI DEĞERLERİ KALDIRMA**

```python
df.sample(5)
```

```python
sns.boxplot(y=df.Views);
plt.title("Views Box Plot ");
plt.show();
```

```python
Q1_views = np.percentile(df.Views,25)
Q1_views 
```

```python
Q3_views = np.percentile(df.Views,75)
Q3_views
```

```python
IQR_views = Q3_views-Q1_views
IQR_views
```

```python
#Let's find the upper limit 

upper_limit_views = Q3_views + 1.5*IQR_views
upper_limit_views 
```

```python
#Let's find a lower limit.

lower_limit_views = Q1_views - 1.5*IQR_views
lower_limit_views
```

```python
outliers_views = df[(upper_limit_views<df.Views) | (lower_limit_views>df.Views)].index

print(outliers_views)
```

```python
print("Upper Limit:", upper_limit_views)
print("Lower Limit:", lower_limit_views)
print("Outliers:", outliers_views)
```

```python
outliers_views
```

```python
df = df[~((df['Views'] < lower_limit_views) | (df['Views'] > upper_limit_views))]
```

```python
sns.boxplot(x=df.Views);
```

```python
df.groupby("Track")["Tempo"].mean().sort_values( ascending=False).reset_index().head(10)
```

```python
# Verileri gruplayın, ortalama alın ve sıralayın
grouped_data = df.groupby("Track")["Tempo"].mean().sort_values( ascending=False).reset_index().head(10)

# Çubuk grafik çizin
plt.figure(figsize=(12, 6))
plt.barh(grouped_data["Track"], grouped_data["Tempo"], color='skyblue')
plt.xlabel('Tempo')
plt.ylabel('Track')
plt.title('Tempo Değerine Göre En Hızlı 10 Şarkı')
plt.gca().invert_yaxis()  # Ters sıralama yapın, en yüksek tempodan en düşüğe doğru sıralansın.
plt.show()
```

Şarkının ortalama tempo değerleri analizi, müziğin ritmi ve tempo özelliklerini incelemek için kullanılan bir yöntemdir. Bu analiz, şarkıların hızını ve ritmini ölçmek amacıyla tempo (bpm - vuruçuk başına darbe) değerlerini kullanır. Şarkıların ortalama tempo değerleri, müziğin dinleyiciler üzerindeki etkisini ve müzik tarzlarını anlamamıza yardımcı olur.

# Spotify Şarkılarının Toplam Stream, Likes, Views ve Comments Sayıları (En İyi 10)

```python
# Trackları Streamlara göre gruplayalım sonra atama yapalım.ilk 10 satırı alalım.
grouped_stream = df.groupby("Track")["Stream"].sum().sort_values(ascending=False).reset_index().head(10)
grouped_likes = df.groupby("Track")["Likes"].sum().sort_values(ascending=False).reset_index().head(10)
grouped_views = df.groupby("Track")["Views"].sum().sort_values(ascending=False).reset_index().head(10)
grouped_comments = df.groupby("Track")["Comments"].sum().sort_values(ascending=False).reset_index().head(10)

# çubuk grafikleri oluşturalım.
plt.figure(figsize=(15, 10))

# Stream bar grafiği
plt.subplot(2, 2, 1)
sns.barplot(x='Stream', y='Track', data=grouped_stream, palette='Blues_r')
plt.title('Top 10 Tracks by Total Stream')
plt.xlabel('Total Stream')

# Likes bar grafiği 
plt.subplot(2, 2, 2)
sns.barplot(x='Likes', y='Track', data=grouped_likes, palette='Greens_r')
plt.title('Top 10 Tracks by Total Likes')
plt.xlabel('Total Likes')

# Views  bar grafiği
plt.subplot(2, 2, 3)
sns.barplot(x='Views', y='Track', data=grouped_views, palette='Reds_r')
plt.title('Top 10 Tracks by Total Views')
plt.xlabel('Total Views')

# Comments  bar grafiği
plt.subplot(2, 2, 4)
sns.barplot(x='Comments', y='Track', data=grouped_comments, palette='Blues_r')
plt.title('Top 10 Tracks by Total Comments')
plt.xlabel('Total Comments')

plt.tight_layout()
plt.show()
```

Spotify'da en iyi 10 şarkının toplam stream, likes, views ve comments sayılarına yönelik analiz, müziğin dijital platformlardaki etkisini gösteriyor. Bu şarkılar, genellikle büyük bir dinleyici kitlesi tarafından takip ediliyor ve popülerliğini kanıtlamış durumda.

Toplam stream sayıları, bu şarkıların ne kadar sık dinlendiğini ve müzik platformunda ne kadar çok tercih edildiğini gösteriyor. Likes sayısı, dinleyicilerin bu şarkıları ne kadar beğendiğini ve topluluk tarafından olumlu bir şekilde karşılandığını yansıtıyor. Views, bu şarkıların video platformlarında ne kadar izlendiğini ve müziğin görsel bileşenlerinin etkisini gösteriyor. Comments sayısı ise dinleyicilerin bu şarkılar hakkında ne kadar çok yorum yaptığını ve müziğe katkı sağladığını gösteriyor.

Bu analiz, müziğin dijital dünyadaki etkisini ölçmek ve popüler şarkıları belirlemek için kullanılıyor. Ayrıca, dinleyici tercihlerini anlama, sanatçıların başarılarını değerlendirme ve müzik endüstrisindeki trendleri izleme açısından önemli bir araçtır. En iyi 10 şarkının bu metriklerdeki yüksek başarıları, müziğin dijital çağdaki gücünü ve etkisini vurguluyor.

# Lisanslı İçeriklerin ve Resmi Videoların Görüntülenmelerinin Karşılaştırılması

```python
# lisanslı içeriklerin görüntülenme ortalaması
licensed_avg_views = df[df['Licensed'] == 1]['Views'].mean()

# resmi videoların görüntülenme ortalaması
official_avg_views = df[df['Official Video'] == 1]['Views'].mean()

print(f"Lisanslı içeriklerin görüntülenme ortalaması: {licensed_avg_views:.0f}")
print(f"Resmi videoların görüntülenme ortalaması: {official_avg_views:.0f}")

# karşılaştırmayı yapalım.
if licensed_avg_views > official_avg_views:
    print("Lisanslı içerikler, resmi videolardan daha popüler.")
elif licensed_avg_views < official_avg_views:
    print("Resmi videolar, lisanslı içeriklerden daha popüler.")
else:
    print("Lisanslı içerikler ve resmi videolar aynı popülerliğe sahip.")
```
